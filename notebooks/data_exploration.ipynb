```python
import pandas as pd

df = pd.read_csv("../data/raw/Startup Failure (Finance and Insurance).csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df["failure"] = 1
```

## Convert years of operation

```python
df["Years of Operation"] = df["Years of Operation"].str.extract('(\d+)').astype(float)
```

## Convert funding

```python
df["How Much They Raised"] = df["How Much They Raised"].str.replace('[^0-9]', '', regex=True)
df["How Much They Raised"] = pd.to_numeric(df["How Much They Raised"], errors='coerce')
```

```python
df = df.drop(columns=[
    "Name",
    "What They Did",
    "Why They Failed",
    "Takeaway"
])
```

```python
df = df.fillna(0)
```

```python
import numpy as np

df_success = df.copy()
df_success["failure"] = 0

for col in df.select_dtypes(include=[np.number]).columns:
    if col != "failure":
        df_success[col] = df_success[col] * np.random.uniform(0.3, 0.9) + np.random.normal(0, 0.1)

df_combined = pd.concat([df, df_success])
```

```python
df_combined["startup_id"] = range(len(df_combined))
```

```python
time_data = []

for _, row in df_combined.iterrows():
    for week in range(1, 13):  # 12 weeks
        temp = row.copy()
        temp["week"] = week
        
        # simulate risk growth
        for col in df_combined.select_dtypes(include=[np.number]).columns:
            if col not in ["failure", "startup_id"]:
                temp[col] = temp[col] * (1 + 0.05 * week)
        
        time_data.append(temp)



df_final = pd.DataFrame(time_data)
```

```python
df_final.to_csv("../data/processed/final_dataset.csv", index=False)
```
